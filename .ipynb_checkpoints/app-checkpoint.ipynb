# 3-Scale the data

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
selected_features =['bedrooms','bathrooms','sqft_living', 'sqft_lot', 'floors', 'sqft_above','sqft_basement']
```

```python
X =house_df[selected_features]
y =house_df['price']
```

```python
scaler =MinMaxScaler()
```

```python
X_scaled =scaler.fit_transform(X)
```

```python
# reshape y
y =y.values.reshape(-1, 1)
```

```python
y_scaled =scaler.fit_transform(y)
```

```python
print(X_scaled.shape)
print(y_scaled.shape)
```

```python
print(scaler.data_max_)
print(scaler.data_min_)
```

# 4-Train The Model

```python
X_train, X_test, y_train, y_test =train_test_split(X_scaled, y_scaled, test_size=0.25)
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model =Sequential()
```

```python
model.add(Dense(50, input_dim=7, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error')
```

```python
epochs_hist =model.fit(X_train, y_train, epochs=100, batch_size=50, validation_split=0.2)
```

# 5-Evaluate The Model

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training & Validation Loss')
plt.legend(['Training Loss','Validation Loss'])
```

```python
# Testing
# bedrooms, bathrooms, sqft_living, sqft_lot, floors, sqft_above, sqft_basement  
X_test_1 =np.array([[4, 3, 1960, 5000, 1, 2000, 3000]])

scaler_1 =MinMaxScaler()

X_test_1_scaled =scaler_1.fit_transform(X_test_1)
```

```python
y_predict_1 =model.predict(X_test_1_scaled)
```

```python
y_predict_1
```

```python
scaler.inverse_transform(y_predict_1)
```

```python
y_predict =model.predict(X_test)
plt.plot(y_test, y_predict, '^', color='r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
```

```python
y_predict_original =scaler.inverse_transform(y_predict)
y_test_original =scaler.inverse_transform(y_test)
```

```python
y_predict =model.predict(X_test)
plt.plot(y_test_original, y_predict_original, '^', color='r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.xlim(0, 5000000)
plt.ylim(0, 3000000)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
```

```python
k =X_test.shape[1]
n =len(X_test)

RMSE =float(format(np.sqrt(mean_squared_error(y_test_original, y_predict_original)), '.3f'))
MSE =mean_squared_error(y_test_original, y_predict_original)
MAE =mean_absolute_error(y_test_original, y_predict_original)
r2 =r2_score(y_test_original, y_predict_original)
r2_adjusted =1-(1-r2) * (n-1) / (n-k-1)
```

```python
print('RMSE = ', RMSE)
print('MSE = ', MSE)
print('MAE = ', MAE)
print('r2 = ', r2)
print('r2_adjusted = ', r2_adjusted)
```

# 7 - Model Improvment With more features

##### let's increase the number of features ( Independent variables )

```python
selected_features =['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']
```

```python
X =house_df[selected_features]
y =house_df['price']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
```

```python
X_scaled =scaler.fit_transform(X)
```

```python
y =y.values.reshape(-1, 1)
y_scaled =scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X_scaled, y_scaled, test_size=0.25)
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model =Sequential()
```

```python
model.add(Dense(50, input_dim=len(selected_features), activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam', loss='mean_squared_error' )
```

```python
epochs_hist =model.fit(X_train, y_train, epochs=100, batch_size=50, verbose=1, validation_split=0.2 )
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training & Validation Loss')
plt.legend(['Training Loss','Validation Loss'])
```

```python
y_predict =model.predict(X_test)
plt.plot(y_test, y_predict, '^', color='r')
plt.xlabel('Model Predictions')
plt.ylabel('True Value (ground truth)')
plt.title('Linear Regression Predictions')
plt.show()
```

```python
y_predict_original =scaler.inverse_transform(y_predict)
y_test_original =scaler.inverse_transform(y_test)
```

```python
k =X_test.shape[1]
n =len(X_test)

RMSE =float(format(np.sqrt(mean_squared_error(y_test_original, y_predict_original)), '.3f'))
MSE =mean_squared_error(y_test_original, y_predict_original)
MAE =mean_absolute_error(y_test_original, y_predict_original)
r2 =r2_score(y_test_original, y_predict_original)
r2_adjusted =1-(1-r2) * (n-1) / (n-k-1)
```

```python
print('RMSE = ', RMSE)
print('MSE = ', MSE)
print('MAE = ', MAE)
print('r2 = ', r2)
print('r2_adjusted = ', r2_adjusted)
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Model Loss progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training & Validation Loss')
plt.legend(['Training Loss','Validation Loss'])
```
